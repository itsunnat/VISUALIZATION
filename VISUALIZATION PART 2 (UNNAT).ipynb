```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.info()
```

```python
sns.displot(titanic["age"])
plt.show()
```

```python
#Chart showing age wise distribution and class wise division in each group

sns.displot(titanic,x="age",hue="class")
plt.show()
```

```python
#Chart showing age wise distribution and Sex wise division in each group

sns.displot(titanic,x="age",hue="sex",palette="rocket")
plt.show()
```

```python
#Chart showing age wise distribution and Survived wise division in each group

sns.displot(titanic,x="age",hue="survived",palette="Reds")
plt.show()
```

```python
#Chart showing age wise distribution and embarked wise division in each group

sns.displot(titanic,x="age",hue="embarked")
plt.show()
```
